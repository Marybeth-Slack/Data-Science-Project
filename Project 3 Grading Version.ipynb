## Cleaning Data

The code below uses the .isnull function to search for rows that do not have values in some of the cells, and then transform them into a boolean series, which will place all the null values as false.  I refrecned my dataframe "df" and then the column I wanted to look for, in this case CollectionDate and PropID (The details of why I chose these two are listed below the code).

I then made a new dataframe that had now been cleaned, called cleaned_df. I used .loc to locate the newly made columns now with null data and used the tilda(~) to add to not add it, so only things that are not null are in the cleaned dataframe. 

I then printed out the new cleaned dataframe 

```python
new_dates = pd.isnull(df['CollectionDate'])
new_propID = pd.isnull(df['PropID'])

cleaned_df = df.loc[~new_dates]
cleaned_df = df.loc[~new_propID]

cleaned_df
```

To clean the data I went through the datafiles to check if the data was consistent or missing any values. Once its downloaded as an excel sheet it was obvious that many cells were empty in the collection dates, and propID columns. 

However to confirm there wasn't any other empty cells, I selected each column and used the search function,( ctrl, f) to see if there were any other columns that had empty cells. 

Collection dates, and propID, both were the only ones with empty cells. I cleaned the data using .isnull and .loc, then I ran and looked through the data to see if there were any errors or unexpected outputs that might effect how it runs. 

There were none so I decided the data was sufficently cleaned.
