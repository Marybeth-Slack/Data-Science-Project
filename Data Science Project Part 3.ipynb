## Section 1: Data and research questions
This dataset is taken from the paper: Range maps and waterbody occupancy data for 1158 freshwater macroinvertebrate genera in the contiguous USA

It can be found at: https://www.nature.com/articles/s41597-024-03845-5#Sec7

This paper was written to make up for a lack of information regarding freshwater macroinvertebrates represented in range maps. Range maps are important to maintain biodiversity and essential information on monitoring species.
 
To increase our understanding and awareness of these species the authors created range maps from 11,628 sites containing 1158 genera of freshwater macroinvertebrates and relevant data based on:
	20 years of data across the US from 11,628 sites and 6,906,990 organisms
	25 different freshwater macroinvertebrates classes and corresponding data
	Relevant data specific for each site.


The observations for this study are the 11,628 sites which had data taken for freshwater macroinvertebrates. There is a total of 15 features for the dataset being used for the sites, and they include: latitude, longitude, site number, collection time, collection year, visit number, area sample, identification, genus, program, habitat, family, order, class, and phylum.

The other datasets in this study include maps for each genus which shows range, occurrence, and non-detected rate, a dataset including geographic specific data and taxonomic classification, the data for numbers of waterbody types, the code that was used to make some of the maps, a dataset with all of the genera classifications, and the metadata contained in a single document.

The research question for this study was if they could address the knowledge gap between freshwater vertebrates and freshwater macroinvertebrates by creating range maps to understand the range of species across the US while producing data on the biodiversity and the current climate associated with macroinvertebrates. They hypothesized that they could use collected data to make range maps. Which include creating maps of 11,628 sites based on data from the USEPA National Aquatic Resource Surveys, which included 6,906,990 organisms. The authors used clustering algorithms on occurrences at the sites and then created layered polygons to create maps representing 25 different classes of macroinvertebrate. They also included occupancy of waterbody type and the corresponding location. 

A different research question that I would ask would be, about the difference of freshwater vertebrates to macroinvertebrates in terms of water habitation. ## Cleaning Data

```python
new_dates = pd.isnull(df['CollectionDate'])
new_propID = pd.isnull(df['PropID'])

cleaned_df = df.loc[~new_dates]
cleaned_df = df.loc[~new_propID]

cleaned_df
```

To clean the data I went through the datafiles to check if the data was consistent or missing any values. I found that two columns: collection dates, and propID, both had empty cells. To fix this I used the .isnull function on both columns to check for missing values and then transform them into a boolean series, which will place all the null values as false. 

We then can use .loc on both columns to find the elements that have been designated "true" and then return the correct value for each. In summary this gives us our cleaned data with all empty cells taken out.
